```python
###  Z_ score
```

```python
from scipy import stats
import numpy as np
```

```python
z = np.abs(stats.zscore(boston_df))
```

```python
z
```

```python
threshold=3
np.where(z>threshold)
#its not the actual score its list of rows and columns respectievely for which the z is greater than 3
```

```python
z.iloc[55][1]
```

```python
## IQR
```

```python
Q1 = boston_df.quantile(0.25)
Q2 = boston_df.quantile(0.75)
IQR = Q2-Q1
```

```python
IQR
```

```python
np.where((boston_df < (Q1 - 1.5*IQR)) | (boston_df > (Q2+1.5*IQR)))
```

```python
boston_df1=boston_df[(z<3).all(axis=1)] ###selecting the ros without outlier
```

```python
boston_df
```

```python
boston_df1
```

```python
boston_df2=boston_df.copy()
```

```python
for i,j in zip(np.where(z>threshold)[0], np.where(z>threshold)[1]):
    boston_df2[i,j] = boston_df.iloc[:,j].median()           
               
```

```python
z = np.abs(stats.zscore(boston_df2))
np.where(z>3)
```
