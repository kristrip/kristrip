```python
import scipy.stats as stats
```

```python
h = np.asarray(Data['GRE Score'])
```

```python
h
```

```python
h = sorted(h)
```

```python
h
```

```python
len(h)
```

```python
fit = stats.norm.pdf(h,np.mean(h),np.std(h))
```

```python
len(fit)
```

```python
plt.plot(h,fit,'-',linewidth=2,label="NormalDistributiom with same mean and varience")
plt.hist(h,normed=True,bins=100,label="Actual Distribution")
plt.legend()
plt.show()
```
